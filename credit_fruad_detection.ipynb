# Credit Fraud Detection 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Check the Data

```python
credit_df=pd.read_csv("creditcard.csv")
credit_df.head(10)
```

```python

credit_df.describe()
```

# Data is not Balanced

```python
credit_df['Class'].value_counts()
```

# Data is not Null Value

```python
credit_df.info()
```

# Check the leg_Data or Fruad_Data

```python
leg_data=credit_df[credit_df.Class==0]
fruad_data=credit_df[credit_df.Class==1]

print("leg_Data",leg_data.shape)

print("fruad_data",fruad_data.shape)
```

# Data is Balance

```python
credit_leg_equal=credit_df.sample(n=492)
credit_card=pd.concat([credit_leg_equal,fruad_data],axis=0)


credit_card['Class'].value_counts()

```

```python
credit_card.groupby('Class').mean()
```

```python
# Step-1:
# X is Features and Y is Target
X = credit_card.drop('Class', axis=1)
Y = credit_card['Class']

# Step-2:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
model = LogisticRegression()
model.fit(X_train, Y_train)

# Step-3:
train_accuracy = model.score(X_train, Y_train)
print(f"Accuracy score for Train Model: = {train_accuracy:.2f}")

test_accuracy = model.score(X_test, Y_test)
print(f"Accuracy score for Test Model: = {test_accuracy:.2f}")

# Step-4:
try:
    user_input = input("Enter credit values separated by commas: ")
    input_values = list(map(float, user_input.split(',')))  # Split and convert to float
    prediction = model.predict([input_values])  # Predict
    if prediction == [0]:
        print("Legitimate transaction")
    else:
        print("Fraud detection")
except ValueError as er:
    print(f"Invalid input: {er}")
```

```python
plt.figure(figsize=(4, 3)) 
sentiment_counts = credit_card['Class'].value_counts()
sentiment_labels = ['legitimate data', 'fraud data'] 
plt.bar(sentiment_labels, sentiment_counts, color=['green', 'red'])
plt.title('Credit-Fraud-Detection')
plt.xlabel('Bar-chart of Fraud and Legitimate') 
plt.ylabel('Count-value')
plt.xticks(rotation=0) # Keep x labels horizontal # Show the chart
plt.show()
```
